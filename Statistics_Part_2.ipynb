## Theory

1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It involves comparing a null hypothesis to an alternative hypothesis and determining the likelihood of the observed data under the null hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- Null hypothesis (H₀): It is the hypothesis that there is no effect or no difference, often representing a default position.
- Alternative hypothesis (H₁ or Ha): It is the hypothesis that there is an effect or a difference, which contradicts the null hypothesis.
The null hypothesis is typically tested to see if there is sufficient evidence to reject it in favor of the alternative hypothesis.

3. What is the significance level in hypothesis testing, and why is it important?
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It is typically set at 0.05, meaning there is a 5% chance of incorrectly rejecting the null hypothesis. The significance level helps control the likelihood of Type I errors and is essential for determining the threshold for statistical significance.

4. What does a P-value represent in hypothesis testing?
The P-value represents the probability of obtaining results at least as extreme as those observed in the sample, assuming the null hypothesis is true. A lower P-value indicates stronger evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
- If P ≤ α: There is enough evidence to reject the null hypothesis.
- If P > α: There is not enough evidence to reject the null hypothesis.
In practice, a P-value of 0.05 (or 5%) is commonly used as the cutoff for statistical significance.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type I error: Occurs when the null hypothesis is incorrectly rejected (false positive).
- Type II error: Occurs when the null hypothesis is incorrectly not rejected (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-tailed test: Tests for the possibility of an effect in one direction (either greater than or less than a certain value).
- Two-tailed test: Tests for the possibility of an effect in both directions (either greater than or less than a certain value).
A two-tailed test is more commonly used when no specific direction of the effect is hypothesized.

8. What is the Z-test, and when is it used in hypothesis testing?
The Z-test is used when the sample size is large (typically n > 30), or when the population standard deviation is known. It compares the sample mean to the population mean, assuming the data is normally distributed.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
The Z-score is calculated as:
Z = (X - μ) / (σ / √n)
Where:
- X = sample mean
- μ = population mean
- σ = population standard deviation
- n = sample size
The Z-score represents how many standard deviations the sample mean is away from the population mean. It is used to determine the significance of the result in hypothesis testing.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is used instead of the normal distribution when the sample size is small (n < 30) and the population standard deviation is unknown. It accounts for the additional variability in small samples.

11. What is the difference between a Z-test and a T-test?
- Z-test: Used when the sample size is large or the population standard deviation is known.
- T-test: Used when the sample size is small or the population standard deviation is unknown.

12. What is the T-test, and how is it used in hypothesis testing?
The T-test is used to compare the means of two groups or a sample mean to a population mean when the sample size is small and the population standard deviation is unknown. There are different types of T-tests (one-sample, independent, paired).

13. What is the relationship between Z-test and T-test in hypothesis testing?
Both tests are used for comparing sample statistics to population parameters, but the T-test is used when the population standard deviation is unknown, and the sample size is small, while the Z-test is used when the population standard deviation is known or the sample size is large.

14. What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval provides a range of values that is likely to contain the true population parameter with a certain level of confidence (e.g., 95%). It is used to express the uncertainty in an estimate.

15. What is the margin of error, and how does it affect the confidence interval?
The margin of error represents the range within which the true population parameter is likely to lie, given the sample data. It affects the width of the confidence interval—larger margins of error result in wider intervals.

16. How is Bayes' Theorem used in statistics, and what is its significance?
Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It allows for the calculation of posterior probabilities, combining prior beliefs with observed data.

17. What is the Chi-square distribution, and when is it used?
The Chi-square distribution is used for categorical data and is commonly applied in tests of independence, goodness-of-fit tests, and tests of homogeneity. It is used when the data is divided into categories.

18. What is the Chi-square goodness of fit test, and how is it applied?
The Chi-square goodness of fit test is used to determine if a sample data fits a population with a specific distribution. It compares the observed frequencies of categories to the expected frequencies under a specific hypothesis.

19. What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is used in analysis of variance (ANOVA) and tests for equality of variances. It is the ratio of two independent chi-square variables divided by their respective degrees of freedom.

20. What is an ANOVA test, and what are its assumptions?
The ANOVA test (Analysis of Variance) compares means across multiple groups. The main assumptions are:
- Independence of observations
- Normality of data within each group
- Homogeneity of variances (equal variances across groups)

21. What are the different types of ANOVA tests?
- One-way ANOVA: Compares the means of three or more independent groups based on one factor.
- Two-way ANOVA: Compares the means across groups with two factors (and their interaction).
- Repeated measures ANOVA: Compares the means of related groups.

22. What is the F-test, and how does it relate to hypothesis testing?
The F-test is used to compare the variances of two or more groups to determine if they are significantly different. It is commonly used in ANOVA to test if group means are different based on the assumption that the variances of the groups are equal.

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import matplotlib.pyplot as plt

z = 2.1
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z, color='black', linestyle='--', label=f'Z = {z}')
plt.title("Two-tailed Z-test Decision Region")
plt.legend()
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
def visualize_errors(mu0, mu1, sigma, n, alpha):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    se = sigma / np.sqrt(n)
    critical_value = norm.ppf(1 - alpha)
    critical_point = mu0 + critical_value * se

    plt.plot(x, norm.pdf(x, mu0, se), label="H0 distribution")
    plt.plot(x, norm.pdf(x, mu1, se), label="H1 distribution")
    plt.fill_between(x, norm.pdf(x, mu0, se), where=x > critical_point, color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, norm.pdf(x, mu1, se), where=x < critical_point, color='blue', alpha=0.3, label="Type II Error")
    plt.axvline(critical_point, color='black', linestyle='--', label="Critical Value")
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

visualize_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
from scipy import stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = stats.t.ppf((1 + confidence) / 2, len(data) - 1) * sem
    return mean - margin, mean + margin

data = np.random.normal(100, 15, 50)
interval = confidence_interval(data)

print(f"95% Confidence Interval: {interval}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    prob_not_A = 1 - prior_A
    prob_B = prob_B_given_A * prior_A + prob_B_given_not_A * prob_not_A
    posterior_A_given_B = (prob_B_given_A * prior_A) / prob_B
    return posterior_A_given_B

posterior = bayes_theorem(0.01, 0.95, 0.05)
print(f"Posterior probability: {posterior}")
```
